# working with combinations and permutation 
- combinations  -> ordered does not matter
- permutations  -> order does matter

```python
import itertools #importing itertools because it has combinations and permutations method
```

```python
list1 = [1,2,3]
combinations = itertools.combinations(list1, 3)
for c in combinations:
    print(c)
#orders does not matter thats why it gives output only 1 because even if we reverse thet output it will be same
```

```python
permutations = itertools.permutations(list1, 3)
for p in permutations:
    print(p)
#order does matter even here (1,2,3) and (3,2,1) contains same number but still it prints out both of them
```

```python
list1 = [1,2,3]
combinations = itertools.combinations(list1, 2)
for c in combinations:
    print(c)
#orders does not matter
```

```python
permutations = itertools.permutations(list1, 2)
for p in permutations:
    print(p)
```

```python
#another example
#gives all the sum of the list is 10
import itertools
list2 = [1,2,3,4,5,6]
combinations = itertools.combinations(list2, 3)
Permutations = itertools.permutations(list2, 3)
print('for combinations')
print([result for result in combinations if sum(result) ==10]) #for combinations

#in combinations we get 3 result only because order does not matter here
print("for permutations")
print([result for result in Permutations if sum(result) ==10]) #for permutations
```

```python
#using combinations
import itertools
words = 'sample'
letters1 = 'plmeas'
combinations = itertools.combinations(words, 6)
permutations = itertools.permutations(words, 6)
for c in combinations:
    if "".join(c) == letters1:
        print('match found')
        break
else:
    print("no match found")
```
